```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('ActivationDataset.csv')  
x = df['input'].values 
```

```python
df
```

```python
def plot(func, yaxis=(-1.4, 1.4)):
    plt.ylim(yaxis)
    plt.axhline(lw=1, c='black')
    plt.axvline(lw=1, c='black')
    plt.grid(alpha=0.4, ls='-.')
    plt.plot(x, func(x), 'r-')
    plt.show()
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
plt.title('Sigmoid')
plot(sigmoid, yaxis=(-0.4, 1.4))
```

```python
def tanh(x):
    return (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))
```

```python
plt.title('Tanh')
plot(tanh)
```

```python
def relu(x):
    return np.maximum(0, x)
```

```python
plt.title('ReLu')
plot(relu)
```

```python
def leaky_relu(x):
    return np.maximum(0.1 * x, x)
```

```python
plt.title('Leaky_ReLu')
plot(leaky_relu)
```
